```python
# Вижив - не вижив (скільки осіб вижило, а скільки загинуло); Вижив по класах; Вижив за віком (молоді до 30 років,
# середні до 60, старші > 60); за походженням (знайти звідки найбільше було пасажирів); як плата за квиток залежить від
# класу каюти (використати boxplot).


import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv('titanic.csv')
# pd.set_option('display.max_columns', None)
# pd.set_option('display.max_rows', None)
```

```python
df.groupby('Survived').size().plot(kind='bar', subplots=True, figsize=(8, 8), color=['rosybrown', 'paleturquoise'])
plt.title('Bar of survivors')
plt.ylabel('Amount of people')
plt.show()
```

```python
df.loc[df['Survived'] == 1].groupby('Pclass').size().plot(kind='pie', subplots=True, figsize=(8, 8),
                                                          colors=['powderblue', 'bisque', 'thistle'],autopct='%1.2f%%',
                                                          shadow=True, startangle=90)
plt.title('Pie Chart of survivors by class')
plt.ylabel('')
plt.show()
```

```python
df["Age"] = df["Age"].fillna(0)
df['AgeGroup'] = pd.cut(df["Age"], [-1, 0, 60, 150], labels = ['junior', 'middle', 'senior'])
sns.barplot(x="AgeGroup", y="Survived", data=df)
plt.show()
```

```python
df.groupby('Embarked').size().plot(kind='bar', subplots=True, figsize=(8, 8), color=['mediumturquoise', 'gold', 'salmon'])
plt.title('Bar Chart of embarked')
plt.ylabel('Amount of people')
plt.show()
```

```python
sns.boxplot(x="Pclass", y="Fare", data=df)
```
